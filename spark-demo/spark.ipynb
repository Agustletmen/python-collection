SparkSession：PySpark 2.0+ 的核心入口，用于创建 DataFrame、执行 SQL 等，替代旧版的 SparkContext。
DataFrame：分布式数据集合，类似数据库表 / Excel 表格，包含行和列，支持结构化数据处理。
RDD：弹性分布式数据集（Resilient Distributed Dataset），Spark 底层数据结构，适用于非结构化数据（如文本）。
Transformation：转换操作（惰性执行），如 filter、map、groupBy，不会立即执行，仅记录计算逻辑。
Action：行动操作（触发执行），如 show、count、collect，会触发之前的转换操作执行。
