```python
guangdong = {
    "2024": [768000],
    "2023": [739000],
    "2022": [702300],
    "2021": [783000],
}  #   https://www.dxsbb.com/news/47760.html

plt.rcParams["font.family"] = ["SimHei"]  # 
plt.rcParams["axes.unicode_minus"] = False  # 

year = [2021, 2022, 2023]

# sum = [783000, 702300, 739000, 768000]
sum = [783000, 702300, 739000]

wuli = [337988, 399216, 410000]
lishi = [269209, 272196, 290000]

# 
tekongxian_lishi = [548, 532, 540]

# 
benke_lishi = [448, 437, 433]

# 
zhuanke_lishi = [160, 180, 180]

gap = []
for i in range(0, len(wuli)):
    gap.append(wuli[i] - lishi[i])
print(gap)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(year, sum, label="")


axs[0, 1].plot(year, lishi, label="")
axs[0, 1].plot(year, wuli, label="")

axs[1, 0].plot(year, gap, label="")

axs[1, 1].plot(year, tekongxian_lishi, label="")
axs[1, 1].plot(year, benke_lishi, label="")
axs[1, 1].plot(year, zhuanke_lishi, label="")

plt.legend()
plt.show()
```
