```python
# import matplotlib as mpl
import matplotlib.pyplot as plt  # https://www.matplotlib.net/stable/tutorials/introductory/pyplot.html
import numpy as np
import pandas as pd
import scipy as sp
```

```python
guangdong = {
    "2024": [768000],
    "2023": [739000],
    "2022": [702300],
    "2021": [783000],
}  # 广东新课标一卷报考总人数  https://www.dxsbb.com/news/47760.html

plt.rcParams["font.family"] = ["SimHei"]  # 使用黑体
plt.rcParams["axes.unicode_minus"] = False  # 正常显示负号

year = [2021, 2022, 2023]

# sum = [783000, 702300, 739000, 768000]
sum = [783000, 702300, 739000]

wuli = [337988, 399216, 410000]
lishi = [269209, 272196, 290000]

# 特控线
tekongxian_lishi = [548, 532, 540]

# 本科
benke_lishi = [448, 437, 433]

# 专科
zhuanke_lishi = [160, 180, 180]

gap = []
for i in range(0, len(wuli)):
    gap.append(wuli[i] - lishi[i])
print(gap)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(year, sum, label="报名总数")


axs[0, 1].plot(year, lishi, label="历史类")
axs[0, 1].plot(year, wuli, label="物理类")

axs[1, 0].plot(year, gap, label="历史物理类报名人数差")

axs[1, 1].plot(year, tekongxian_lishi, label="特控线（历史）")
axs[1, 1].plot(year, benke_lishi, label="本科线（历史）")
axs[1, 1].plot(year, zhuanke_lishi, label="专科线（历史）")

plt.legend()
plt.show()
```

```python
# 坐标轴刻度显示，不填内容就隐藏
plt.xticks([])
plt.yticks([])

# [xmin, xmax, ymin, ymax]  绘图区域的显示范围
plt.axis([0, 6, 0, 20])
```



1. Figure：红色的外框，可以把理解为一个画板
2. Axes：蓝色的内框，一套坐标轴组合，可以理解为一个坐标系
3. Axis：绿色的横纵坐标轴，上面包含刻度和标签（tick locations和labels）

> 如图2，定义 4 个 subplot，自然就有 4 个Axes

```python
# 子图

plt.figure(figsize=(10, 5))  # 画板

plt.subplot(
    2, 2, 1
)  # figure,(axes1,axes2... axesn) = subplot(图形行数, 图形列数, 第n个, **附加选项)  创建新的 Figure 和一组 Axes
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title("First Subplot")

plt.subplot(3, 2, 2)  # 选择第二个子图
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title("Second Subplot")

plt.show()
```

```python
# 条形图
plt.bar(["A", "B", "C", "D"], [3, 7, 2, 5])
plt.show()

# 饼图
plt.pie([3, 7, 2, 5], labels=["A", "B", "C", "D"])
plt.show()

# 散点图
plt.scatter(x, y)
plt.show()
```
