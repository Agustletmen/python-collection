```python
# 导入核心库
import pandas as pd
import numpy as np

# 模拟MQTT设备上报的时序数据（1000条，1分钟1条）
# 实际场景可替换为读取真实数据（见步骤2）
time_index = pd.date_range(start='2026-02-12 00:00:00', periods=1000, freq='1min')
df = pd.DataFrame({
    'timestamp': time_index,          # 时间戳（核心时序字段）
    'device_id': 'sensor_001',        # 设备ID
    'temperature': np.random.normal(60, 2, 1000),  # 温度（含随机波动）
    'voltage': np.random.normal(220, 1, 1000),     # 电压
    'current': np.random.normal(1.5, 0.1, 1000)    # 电流
})
# 手动插入异常值（模拟设备故障）
df.loc[200:210, 'temperature'] = 80  # 温度突升
df.loc[500:520, 'voltage'] = np.nan  # 电压数据缺失
df.loc[800:810, 'current'] = 5       # 电流异常飙升

# 查看数据前5行（快速验证数据结构）
print("数据前5行：")
print(df.head())
# 查看数据基本信息（行数、列数、数据类型、缺失值）
print("\n数据基本信息：")
print(df.info())
# 查看数值字段的统计指标（均值、标准差、最值）
print("\n统计指标：")
print(df.describe())
```
